```python
df.period_end = pd.to_datetime(df.period_end,format="%Y-%m-%d")
```

```python
grouped = df.groupby(['iCIK','period_end']).count()
grouped = grouped.reset_index()
grouped = grouped.drop(grouped.iloc[:,3:],axis=1)
grouped = grouped.rename(columns = {'date':'holdings'})
grouped.head(5)
```

```python
max(grouped[grouped.iCIK==7195].period_end)-min(grouped[grouped.iCIK==7195].period_end)
```

```python
stime_total = []
for i in np.array(np.unique(grouped.iCIK)):
    stime = max(grouped[grouped.iCIK==i].period_end)-min(grouped[grouped.iCIK==i].period_end)
    stime_total.append(stime.days)
```

```python
average_stime = sum(stime_total) / len(stime_total)
average_stime
```

```python
high_cutoff = average_stime * 2 #High-Survival Fund -> twice the mean survival time 
low_cutoff = average_stime * 0.5 #Low_Survival Fund -> 50% below the mean 
print("High_cutoff is {} and Low_cutoff is {}".format(high_cutoff,low_cutoff))
```

```python
stime_total = np.array(stime_total)
```

```python
print("The number of High-Survival Fund is {}".format(sum(stime_total > high_cutoff)))
print("The number of Mid-Survival Fund is {}".format(sum((stime_total < high_cutoff) & (stime_total > low_cutoff))))
print("The number of Low-Survival Fund is {}".format(sum(stime_total < low_cutoff)))
```

```python
conditions = [stime_total > high_cutoff,(stime_total < high_cutoff) & (stime_total > low_cutoff),stime_total < low_cutoff]
choices = ["High","Mid","Low"]
category = np.select(conditions,choices,default=np.nan)
```

```python
dict1 = {'funds':np.unique(grouped.iCIK),'survival_time':stime_total,'category':category}
```

```python
final = pd.DataFrame(dict1)
final_merged = pd.merge(df,final,left_on="iCIK",right_on="funds",how="left")
```
