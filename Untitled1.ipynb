```python
price_use = prices[['stock_id','close']]
price_use
```

```python
price_use2 = price_use.copy()
price_use2['date'] =price_use.index-pd.DateOffset(months = 3)
price_use2 = price_use2.set_index('date')
price_use2.columns = ['stock_id','3MA close']
price_new = pd.merge(price_use,price_use2, 'right', on = ['date', 'stock_id'])
price_new
```

```python
price_new['return'] = (price_new['3MA close']-price_new['close'])/(price_new['close'])
price_new
```

```python
price_new = price_new[['stock_id','return']]
price_new = price_new.dropna()
```

```python
price_table = pd.pivot_table(price_new, values = 'return',index = ['date'],columns = 'stock_id',aggfunc=np.mean)
price_table
```

```python
price_mean = price_table.mean()
price_mean
```

```python
price_mean = price_mean.sort_values(ascending = True)
price_mean
```

```python
MEAN = price_mean.mean()
```

```python
high_perf = price_mean[price_mean >= MEAN*1.2]
low_perf = price_mean[price_mean <= MEAN*0.8]
med_perf = price_mean[price_mean >= MEAN*0.8]
med_perf = med_perf[med_perf <= MEAN*1.2]
# med_perf
# high_perf
# low_perf
```

```python
price_vol = price_table.std()
MEAN_vol = price_vol.mean()
price_vol = price_vol.sort_values(ascending = True)
high_vol = price_vol[price_vol >= MEAN_vol*1.2]
low_vol = price_vol[price_vol <= MEAN_vol*0.8]
med_vol = price_vol[price_vol >= MEAN_vol*0.8]
med_vol = med_vol[med_vol <=MEAN_vol*1.2]
MEAN_vol
```
