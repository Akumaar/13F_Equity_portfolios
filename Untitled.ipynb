```python
def turnover_one_fund(df, index):
    tp = df[df['iCIK']==index]
    tp = tp[['stock_id', 'period_end', 'iMARKET_VALUE', 'iLONG_FRACTION']]
    tp['period_end'] = pd.to_datetime(tp['period_end'])
    tp = tp.sort_values(by=['period_end'])
    tp = tp.sort_values(by=['stock_id', 'period_end'])
    tp['buy'] = 0
    stocks = tp['stock_id'].unique()
    for i in stocks:
        temp = tp[tp['stock_id']==i]
        if len(temp) == 1:
            tp.set_value(temp.index[0], 'buy', temp.iloc[0]['iMARKET_VALUE'])
        else:
            diff = np.diff(temp['iMARKET_VALUE'])
            diff = np.insert(diff, 0, temp.iloc[0]['iMARKET_VALUE'])
            for j in range(len(temp)):
                tp.set_value(temp.index[j],'buy',diff[j])
    tp = tp.sort_values(by='period_end')
    temp = tp['buy'].values
    zeros = np.zeros_like(temp)
    tp['sell'] = np.maximum(zeros, -temp)
    tp['buy'] = np.maximum(zeros, temp)
    grouped = tp.groupby(['period_end']).sum()
    grouped['turnover'] = np.minimum(grouped['buy'], grouped['sell'])/grouped['iMARKET_VALUE']
    grouped = grouped['turnover']
    grouped = grouped.reset_index()
    grouped['iCIK'] = index
    return grouped
```

```python
result = []
for i in df['iCIK'].unique():
    result.append(turnover_one_fund(df, i))
```

```python
result = pd.concat(result)
```

```python
median = result.groupby('period_end').median()
```

```python
result = result[['period_end', 'turnover', 'iCIK']].sort_values(by='period_end')
result
```

```python
f = lambda row : row['turnover'] > median.loc[row['period_end']]['turnover']
```

```python
result['greater'] = result.apply(f, axis = 1)
```
