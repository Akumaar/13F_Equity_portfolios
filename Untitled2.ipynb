```python
import pandas as pd
import numpy as np 
import datetime
```

```python
prices = pd.read_csv('prices.csv')
factors = pd.read_csv('13f.csv')
```

```python
prices.columns = ['stock_id','exchange','ticker','date','volume','open','high','low','close']
factors.columns = ['iRECORD_ID','iCIK','stock_id','iPERIOD_END','date','iAMEND','iRESTATEMENT','iTYPE','iQTY','iMARKET_VALUE','iLONG_FRACTION']
```

```python
prices_copy = prices.copy()
factors_copy = factors.copy()
```

```python
factors = factors[factors['iAMEND']== 0]
factors = factors[factors['iRESTATEMENT']== 0]
factors = factors[factors['iTYPE']== 0]
factors
```

```python
factors = factors.drop('iAMEND', axis = 1)
factors = factors.drop('iRESTATEMENT', axis = 1)
factors = factors.drop('iTYPE', axis = 1)
factors
```

```python
factors['gap_day'] = pd.to_datetime(factors['date'])-pd.to_datetime(factors['iPERIOD_END'])
factors = factors.drop('iPERIOD_END', axis = 1)
factors
```

```python
factors = factors[factors['gap_day']<=datetime.timedelta(days = 45)]
factors = factors.drop('gap_day', axis = 1)
factors
```
