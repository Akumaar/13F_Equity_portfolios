```python
import numpy as np
import pandas as pd
import datetime
```

```python
prices = pd.read_csv('prices_from_2013.csv')
prices
```

```python
price_use = prices[['date','stock_id','close']]
price_use
```

```python
price_pivot = pd.pivot_table(price_use, values = 'close',index = ['date'],columns = 'stock_id',aggfunc=np.mean)
price_pivot
```

```python
price_pivot = price_pivot.pct_change()
```

```python
price_pivot
```

```python
price_vol = price_pivot.std()
price_vol
```

```python
price_vol = price_vol.dropna()
price_vol = price_vol[price_vol != 0]
mid_vol = price_vol.quantile([0.25,0.5,0.75])
mid_vol
```

```python
labels = ['high vol','mid vol','low vol']
price_vol = dict(price_vol)
price_vol
```

```python
data_items = price_vol.items()
data_list = list(data_items)
vol_df = pd.DataFrame(data_list)
vol_df.columns = ['stock_id','volatility']
vol_df
```

```python
label = ['low vol','mid vol','high vol']
vol_df['vol group'] = pd.cut(vol_df['volatility'], [0,mid_vol[0.25],mid_vol[0.75],max(vol_df['volatility'])], right = False, labels = label)
vol_df
```

```python
vol_df.to_csv('volatility.csv')
```
