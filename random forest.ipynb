```python
import numpy
import pandas
from sklearn.ensemble import RandomForestRegressor
import datetime
```

```python
factors = pandas.read_csv('Final Merge file.csv')
factors
```

```python
factors['date'] = pandas.to_datetime(factors['date'])
factors = factors.set_index('date')
factors = factors[['stock_id','x21','x1','good fund','irisk','x20','x4','1M Return','2M Return','3M Return']]
factors
```

```python
pred_val = pandas.read_csv('3M return.csv')
pred_val
```

```python
pred_val.columns = ['date','stock_id','pred']
pred_val
```

```python
pred_val['date'] = pandas.to_datetime(pred_val['date'])
pred_val= pred_val.set_index('date')
```

```python
pred_val
```

```python
pred_val.index = pred_val.index - pandas.DateOffset(months = 3)
pred_val
```

```python
pred_table = pandas.pivot_table(pred_val, values = 'pred', index = ['date'],columns = 'stock_id',aggfunc = numpy.sum)
pred_table
```

```python
pred_factor = pred_table.asfreq(freq = '1d', method = 'ffill').stack()
pred_factor
```

```python
pred_factor = pandas.DataFrame(pred_factor)
pred_factor = pred_factor.reset_index('stock_id')
pred_factor
```

```python
factors = pandas.merge(factors, pred_factor, 'right',on = ['date','stock_id'])
factors = factors.dropna()
factors
```

```python
factors.index.unique()
```

```python
factors.columns = ['stock_id','x21','x1','good fund','irisk','x20','x4','1M Return','2M Return','3M Return','return']
train = factors[:'2017,10,01']
test = factors['2017,10,01':]
feat = ['x21','x1','good fund','irisk','x20','x4','1M Return','2M Return','3M Return']
train_feature = train[feat]
train_return = train['return']
test_feature = test[feat]
test_return = test['return']
```

```python
train_feature
```

```python
rf = RandomForestRegressor(n_estimators = 10, random_state = 42)
rf.fit(train_feature,train_return)
test_predict = rf.predict(test_feature)
errors = numpy.mean((test_predict - test_return)**2)
print('Mean Square Error:', round(errors, 5))
```

```python
test['prediction'] = test_predict
test
```

```python
test.to_csv('prediction.csv')
```
